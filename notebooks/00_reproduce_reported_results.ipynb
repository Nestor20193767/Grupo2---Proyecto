# Reproducción de resultados reportados
Este notebook lee únicamente los CSV versionados en `results/metrics/`.


```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
ROOT = Path.cwd()
if not (ROOT / 'results').exists():
    ROOT = ROOT.parent
metrics = ROOT / 'results' / 'metrics'
```

```python
modes = pd.read_csv(metrics / 'classifier_modes.csv')
modes
```

```python
ax = modes.set_index('mode')[['accuracy','recall_macro','f1_macro']].plot(kind='bar', figsize=(9,5))
ax.set_ylabel('Métrica (%)')
ax.set_title('Clasificador TCN por modo')
plt.xticks(rotation=0)
plt.tight_layout()
```
