# Inferencia reproducible con ARRYS
Requiere los artefactos descritos en `models/README.md`.


```python
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
from arrys import ArrysInference, FS
ROOT = Path.cwd()
if not (ROOT / 'models').exists():
    ROOT = ROOT.parent
engine = ArrysInference(ROOT / 'models')
```

```python
beats, z = engine.generate('NSR', n=8, noise=0.8, seed=42)
beats.shape, z.shape
```

```python
time_ms = np.arange(beats.shape[1]) / FS * 1000
for beat in beats:
    plt.plot(time_ms, beat, alpha=0.3)
plt.plot(time_ms, beats.mean(0), linewidth=2)
plt.xlabel('Tiempo (ms)')
plt.ylabel('Amplitud')
plt.title('Latidos NSR sintéticos — seed 42')
plt.show()
```

```python
probabilities, predictions, labels = engine.classify(beats)
labels, probabilities[:2]
```
